```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

# Data Preprocessing and Exploration

```python
data = pd.read_csv("./shark_tank_india(original).csv")
data.head()
```

```python
data.info()
```

```python
# Removing Unsed Columns
data.drop(["Company Website", "Original Air Date", "Episode Title", "Gross Margin", "Net Margin",
          "EBITDA", "Cash Burn", "SKUs", "Has Patents", "Bootstrapped", "Invested Guest Name", 
           "All Guest Names"], axis = 1, inplace = True)
```

```python
# Replacing Null Values with Zero, Not Avialble, and No.
def replaceWithZero(features):
    for column in features:
        data[column].fillna(0, inplace = True)
        data[column].astype(np.float16)

def replaceWithNotAvailable(features):
    for column in features:
        data[column].fillna("Not Available", inplace = True)


zero_features = ["Guest Present", "Ashneer Present", "Peyush Present", "Amit Present", "Aman Present", "Anupam Present", "Vineeta Present",
           "Namita Present", "Guest Debt Amount", "Guest Investment Equity", "Guest Investment Amount",
           "Ashneer Debt Amount", "Ashneer Investment Equity", "Ashneer Investment Amount", "Amit Debt Amount",
           "Amit Investment Equity", "Amit Investment Amount", "Peyush Debt Amount", "Peyush Investment Equity",
           "Peyush Investment Amount", "Aman Debt Amount", "Aman Investment Equity", "Aman Investment Amount",
           "Anupam Debt Amount", "Anupam Investment Equity", "Anupam Investment Amount", "Vineeta Debt Amount",
           "Vineeta Investment Equity", "Vineeta Investment Amount", "Namita Debt Amount", "Namita Investment Equity",
           "Namita Investment Amount", "Advisory Shares Equity", "Royalty Deal", 
            "Number of Sharks in Deal", "Deal Valuation", "Debt Interest", "Total Deal Debt", "Total Deal Equity",
           "Total Deal Amount", "Accepted Offer", "Monthly Sales", "Yearly Revenue", "Couple Presenters", "Transgender Presenters", 
            "Female Presenters", "Male Presenters"]

not_available_features = ['Pitchers City', 'Pitchers State', "Started in"]

replaceWithZero(zero_features)
replaceWithNotAvailable(not_available_features)

data["Deal Has Conditions"].fillna("no", inplace = True)
```

```python
data["Season Start"] = pd.to_datetime(data["Season Start"])
data["Season End"] = pd.to_datetime(data["Season End"])
```

```python
data.to_csv("sharktank_india(clean).csv")
```

# Exporting Data to MySQL Server for Further Analysis

```python
# Establising Connections
engine = create_engine("mysql://root:1234@localhost:3306/SharkTank")
connection = engine.connect()
```

```python
try:
    data.to_sql("dataset", con = connection, index = False, if_exists = 'append')
    print("Data Inserted")
except Exception as e:
    print(f"Error: {str(e)}")
finally:
    connection.commit()
```
